#TITANIC SURVIVAL PREDICTION

Import necessary libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
```

Import the dataset

```python
df = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/Titanic-Dataset.csv')
df.head()
```

Analyze and Visulaize the data

```python
df.info()
```

```python
df.columns
```

```python
target = df[['Survived']]
target.head()
```

we will train model on columns mentioned below

```python
df1 = df[['Pclass','Sex','Age','Fare']]
df1.head()
```

Habdling categorical and continuous columns

```python
df2 = pd.get_dummies(df1.Sex)
DF = pd.concat([df1, df2],axis='columns')
DF.drop('Sex', axis='columns', inplace=True)
DF.head()
```

```python
DF.info()
```

Missing Value Handling

```python
DF_new = DF.fillna(method="ffill")
DF_new
```

```python
DF_new.isna().sum()
```

Test Train splitting the dataset

```python
x_train,x_test,y_train,y_test = train_test_split(DF_new,target, test_size = 0.28, random_state = 1 )
```

```python
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

Model Training

```python
DTC = DecisionTreeClassifier()
DTC.fit(x_train, y_train )
```

```python
DTC.score(x_test, y_test )
```
